```python
#Using object oriented programming concepts in python to create a game of cards:
import random
#Creating a dictionary of values:
values = {"two":2,"three":3,"four":4,"five":5,"six":6,"seven":7,"eight":8,"nine":9,"ten":10,"Jack":11,"Queen":12,"King":13,"Ace":14}
#Creating the list of all the suits:
suits = ("Hearts","Diamonds","Spades","Clubs")
#Creating the list of all the ranks:
ranks = ("two","three","four","five","six","seven","eight","nine","ten","Jack","Queen","King","Ace")

#Creating a card class:
class card:
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
    def __str__(self):
        #Will just return a statement telling us the rank of the suit. Its basically the string representation of the object.
        return "The card is : " + self.rank + " of " + self.suit + " with value " + str(self.value)

#Creating a deck class:

class deck:
    #In the class deck we actually need to create a list of all 52 cards.
    def __init__(self):
        self.all_the_cards = []
        
        for suit in suits:
            for rank in ranks:
                created_card = card(suit,rank)
                self.all_the_cards.append(created_card)
   
    #Method to shuffle the deck :           
    def shuffled_deck_of_cards(self):
        self.shuffled_deck = []
        for card in self.all_the_cards:
            random_card = random.choice(self.all_the_cards)
            self.shuffled_deck.append(random_card)
            
    #Method to pick any random card from the shuffled deck of cards:        
    def pick_the_last_card(self):
        return self.shuffled_deck.pop()
    
    
class player:
    def __init__(self,name):
        self.name = name
        self.all_the_cards = []
    
    
    def remove_cards(self):
        return self.all_the_cards.pop(0)
        
    
    def add_cards(self,new_cards):
        if type(new_cards) == list:
            self.all_the_cards.extend(new_cards)
        else:
            self.all_the_cards.append(new_cards)
    
    def __str__(self):
        return "Player {} has {} cards".format(self.name,len(self.all_the_cards))

            
#Creating the logic for the game:
#Step 1 : To create two instances of player class and one instance of the deck class.

Player1 = player("Hawra")
Player1.all_the_cards
Player2 = player("Ehab")
Player2.all_the_cards

#Step 2: To create a new deck with all the 52 cards:

new_deck = deck()
new_deck.all_the_cards
new_deck.shuffled_deck_of_cards()


#Step 3: To split the shuffled deck equally among the two players:
for cards in range(26):
    card1 = new_deck.pick_the_last_card()
    card2 = new_deck.pick_the_last_card()
    Player1.add_cards(card1)
    Player2.add_cards(card2)
    #print("Player 1 got {}\nplayer 2 got {}\n\n".format(card1, card2))
    #print(*Player1.all_the_cards, sep="\n")
    
#Putting a flag saying that the game is on:
game_on =True


#Counting the number of rounds that the game goes on for :
game_round = 0
#Drawing random card from the pack of cards that each player is having:


while game_on:

    game_round = game_round+1
    #checking if any of the players finished their cards:
    if len(Player1.all_the_cards) == 0 or len(Player2.all_the_cards) == 0:
        game_on = False
        break
        
    
    #the main part of the game :
    #Step1 : The player draws one card from his deck of cards:
    Player_1_cards = []
    Player_1_cards.append(Player1.remove_cards())
    Player_2_cards = []
    Player_2_cards.append(Player2.remove_cards())
    
   
    game_continues = True
    while game_continues:
        
        #Step2 : Now we will compare the card with each other. The player with the higher rank of card will take both the cards:
        if Player_1_cards[-1].value > Player_2_cards[-1].value:
            Player1.add_cards(Player_1_cards)
            Player1.add_cards(Player_2_cards)
            #print("The total number of cards that Player 1 is having is : {}".format(len(Player1.all_the_cards)))
            game_continues = False
    
        elif Player_1_cards[-1].value < Player_2_cards[-1].value:
            Player2.add_cards(Player_1_cards)
            Player2.add_cards(Player_2_cards)
            game_continues = False
        
        
        elif Player_1_cards[-1].value == Player_2_cards[-1].value:
            if len(Player1.all_the_cards) < 4 or len(Player1.all_the_cards) < 4:
                game_on = False
                break
            
            for i in range(4):
                Player_1_cards.append(Player1.remove_cards())
                Player_2_cards.append(Player2.remove_cards())
            
        

print("The number of rounds played : {}".format(game_round))       
if len(Player1.all_the_cards) > len(Player2.all_the_cards):
    print("Player 1 wins")
else:
    print("Player 2 wins")
print("{} \n{}".format(Player1,Player2))
   



        
        
        

    
```
